```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f, color = 'green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

Задание 3.

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
x, y1, y2, y3, y4
```

```python
fig, ax = plt.subplots(nrows= 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])
ax2.plot(x, y2)
ax2.set_title('График y2')
ax3.plot(x, y3)
ax3.set_title('График y3')
ax4.plot(x, y4)
ax4.set_title('График y4')
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
```

Задание 4.

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\tanbe\\Desktop\\creditcard.csv')
```

```python
plt.style.use('fivethirtyeight')
```

```python
class_info = df['Class'].value_counts()
class_info
```

```python
class_info.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

```python
class_info.plot(kind='bar', logy=True)
plt.xticks(rotation=0)
plt.show()
```

```python
fraud = df['Class'] == 1
common = df['Class'] == 0
```

```python
fraud_V1 = df[fraud]['V1']
common_V1 = df[common]['V1']
```

```python
plt.hist(fraud_V1, bins=20, ec='black', color='red', alpha=0.5, density=True, label='Class 1')
plt.hist(common_V1, bins=20, ec='black', color='lightgreen', alpha=0.5, density=True, label='Class 0')
plt.xlabel('Class')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('C:\\Users\\tanbe\\Desktop\\netflix_titles.csv')
```

```python
df.loc[df['release_year'].between(2010, 2016), 'release_year'].value_counts().plot(kind = 'bar', color = 'orange')
plt.xlabel('Год выпуска')
plt.ylabel('Количество фильмов/сериалов')
plt.show()
```

2.

```python
all_films = df['release_year'].value_counts().sort_index()
plt.plot(all_films, linewidth = 1)
plt.show()
```

3.

```python
tv_shows = df.loc[df['type'] == 'TV Show', :]
seasons = tv_shows.duration.value_counts().head(10).reset_index()
```

```python
fig, ax = plt.subplots()
ax.bar(seasons['index'], seasons['duration'])
plt.xticks(rotation=90)
plt.show()
```

4.

```python
ratio_df = df['type'].value_counts().reset_index()
fig, ax = plt.subplots()
ax.pie(ratio_df['type'], labels = ratio_df['index'], autopct='%1.0f%%', explode = [0.2, 0], colors = ['yellow', 'violet'])
plt.show()
```
