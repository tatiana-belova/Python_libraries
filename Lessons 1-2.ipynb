Тема “Работа с данными в Pandas”

Задание 1.

```python
import pandas as pd
```

```python
authors = pd.DataFrame({'author_id': [1, 2, 3],'author_name': ['Тургенев', 'Чехов', 'Островский']})
authors
```

```python
books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3],'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],'price': [450, 300, 350, 500, 450, 370, 290]})
books
```

Задание 2.

```python
authors_price = pd.merge(authors, books, on='author_id', how='inner')
authors_price
```

Задание 3.

```python
top5 = authors_price.nlargest(5,'price')
top5
```

Задание 4.

```python
authors_stat = authors_price.groupby('author_name').agg({'price':['min', 'max', 'mean']})
authors_stat = authors_stat.rename(columns={'min':'min_price', 'max':'max_price', 'mean':'mean_price'})
authors_stat
```

Задание 5.

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = pd.pivot_table(authors_price, values='price', index='author_name', columns='cover', aggfunc='sum', fill_value=0)
book_info
```

```python
book_info.to_pickle('book_info.pkl')
```

```python
book_info2 = pd.read_pickle('book_info.pkl')
```

```python
book_info.equals(book_info2)
```
